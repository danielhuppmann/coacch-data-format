```python
import pyam
```

```python
df = pyam.IamDataFrame(
    "data/COACCH_DataOutput_Example.csv",
    index=["ssp", "rcp", "gcm"],
    region="Resolution"
)
```

```python
df.timeseries()
```

Change unit to pint-compatible convention and convert to million square kilometers.
