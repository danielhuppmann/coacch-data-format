# Working with the COACCH data format for climate change impact analysis

This tutorial notebook shows how to use the pyam package when working with the COACCH data format.

```python
import pyam
```

Import the data as a [**pyam.IamDataFrame**](https://pyam-iamc.readthedocs.io/en/latest/api/iamdataframe.html).

In contrast to the standard IAMC format, the COACCH data format defines a **scenario** by an SSP-RCP-GCM index.

```python
df = pyam.IamDataFrame(
    "data/COACCH_DataOutput_Example.csv", index=["ssp", "rcp", "gcm"], region="Resolution",
)
```

Display timeseries data in wide format.

```python
df.timeseries()
```

Change unit to pint-compatible convention and convert to million square kilometers.

```python
df.convert_unit("1000 ha", "Mha", factor=1, inplace=True)
df.convert_unit("Mha", "million km^2", inplace=True)
```

Show the timeseries with the changed units in wide format.

```python
df.timeseries()
```

Show a simple line plot.

```python
df.plot()
```

As you see above, **pyam** by default doesn't show a legend if there are more than 13 entries.

But we can easily change that behaviour and even specify where to place the legend!

```python
df.plot(legend=dict(loc="outside bottom"))
```
